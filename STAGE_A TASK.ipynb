**In this task, I'm working with the Food Balance Sheet data from the Food and Agriculture Organization (FAO)**

```python
#I would be using just Pandas for this task, hence I'd import only Pandas
import pandas as pd
```

```python
data_df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', sep=',', encoding='latin-1')
data_df
```

QUESTION 13  
WHAT IS THE TOTAL NUMBER AND PERCENTAGE OF MISSING DATA IN 2016 TO 2 DECIMAL PLACES?

```python
#WHAT IS THE TOTAL NUMBER OF MISSING DATA IN 2016?
data_df.isnull().sum()
```

Based on the above code, the total number of missing data in 2016 is 1535

```python
#WHAT IS THE PERCENTAGE OF MISSING DATA IN 2016?
percentage_missing_data = data_df.isnull().sum() * 100 / len(data_df)
percentage_missing_data
```

The percentage of missing data in 2016 to 2 decimal places is 2.52

QUESTION 14  
WHICH YEAR HAD THE HIGHEST CORRELATION WITH "ELEMENT CODE"?

```python
corr_2014 = data_fao["Element Code"].corr(data_fao["Y2014"])
corr_2014
```

```python
corr_2015 = data_fao["Element Code"].corr(data_fao["Y2015"])
corr_2015
```

```python
corr_2016 = data_fao["Element Code"].corr(data_fao["Y2016"])
corr_2016
```

```python
corr_2017 = data_fao["Element Code"].corr(data_fao["Y2017"])
corr_2017
```

```python
corr_2018 = data_fao["Element Code"].corr(data_fao["Y2018"])
corr_2018
```

```python
#creating a list
correlation = [corr_2014, corr_2015, corr_2016, corr_2017, corr_2018]
correlation
```

```python
#WHICH YEAR HAS THE HIGHEST CORRELATION WITH ELEMENT CODE?
max(correlation)
```
